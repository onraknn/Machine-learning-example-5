```python
# x'e 60. sütunu silip ekledik, y'ye ise 60. sütünu atadık
X = sonar_data.drop(columns = 60, axis = 1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

# **Training and test data**

train = eğitim

stratify = katmanlaşma

test_size = 100'de 10u kadar test et demek

```python
# eğitim ve test verilerini atadık
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape) #21 test verimiz var
```

```python
model = LogisticRegression()
```

```python
print(X_train)
print(Y_train)
```

```python
model.fit(X_train, Y_train)
```

```python
 #accuracy on training data
 X_train_predection = model.predict(X_train)
 training_data_accuracy = accuracy_score(X_train_predection, Y_train) # doğruluk değerini burada saklayacağız.
```

```python
print("accuracy on training data : ", training_data_accuracy)
```

```python
 #accuracy on test data
 X_test_predection = model.predict(X_test)
 test_data_accuracy = accuracy_score(X_test_predection, Y_test) # doğruluk değerini burada saklayacağız.
```

```python
print("accuracy on training data : ", test_data_accuracy)
```
